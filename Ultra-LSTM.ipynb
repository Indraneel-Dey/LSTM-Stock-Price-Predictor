```python
import pandas as pd
import tensorflow as tf
import keras
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import LSTM, Dense
import pickle
```

```python
def prepare(df):
    df.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis='columns', inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_axis(df['Date'], inplace=True)
    df = df.dropna()
    data = df['Close'].values.reshape((-1, 1))
    return df, data
```

```python
def model(lookback, data):
    generator = TimeseriesGenerator(data, data, length=lookback, batch_size=20)
    model = Sequential()
    model.add(LSTM(10, activation='relu', input_shape=(lookback, 1)))
    model.add(Dense(1))
    model.compile(loss='mse')
    model.fit(generator, epochs=100, verbose=1)
    return generator, model
```

```python
ultra, ultra_data = prepare(pd.read_csv('ULTRA.csv'))
```

```python
ultra_generator, ultra_model = model(1, ultra_data)
```

```python
ultra_model.save('ultra_model')
pickle.dump(ultra_generator, open('ultra_gen.pkl', 'wb'))
```
